```python
from openai import OpenAI
client = OpenAI(api_key=api_key)

def mensagem_matricula (candidato):
  response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
      {
          "role": "system",
          "content": "Voc√™ √© um especialista em markting educacional."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para {candidato['nome']} sobre a import√¢ncia de realizar a sua matr√≠cula no curso {candidato['curso']}(m√°ximo de 200 caracteres)"
      }
    ]
  )
  return response.choices[0].message.content.strip()

mensagens_geradas = []
for index, candidato_row in candidatos.iterrows():
  mensagem = mensagem_matricula(candidato_row)
  mensagens_geradas.append(mensagem)

candidatos['mensagem'] = mensagens_geradas
```
